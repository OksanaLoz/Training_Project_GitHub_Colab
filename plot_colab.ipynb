```python
import numpy as np
import pylab
#now we plot the graph
x=np.linspace(-1,1,1000)
y=(x)*np.cos(1/x)
pylab.plot(x, y,'m')
pylab.xlabel('x')
pylab.ylabel('y')
pylab.title('y=xcos(1/x)')
pylab.grid()
```
